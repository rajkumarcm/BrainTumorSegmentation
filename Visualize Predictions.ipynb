Load the libraries

```python
import os
from os.path import abspath, join
import nibabel
from matplotlib import pyplot as plt
import numpy as np
os.environ['CUDA_VISIBLE_DEVICES']='-1'
import tensorflow as tf
```

Load the prediction file

```python
tmp_seg = nibabel.load("tmp_pred.nii")
tmp_seg = tmp_seg.get_fdata()
```

```python
tmp_seg.shape
```

Load the Ground Truth segmentation mask

```python
seg_gt = nibabel.load("Brats18_2013_0_1_seg.nii")
seg_gt = seg_gt.get_fdata()
```

```python
# Here tmp_seg referes to the prediction array
```

Load the actual MRI scan image

```python
tmp_img = nibabel.load("Brats18_2013_0_1_t1.nii")
tmp_img = tmp_img.get_fdata().astype(np.uint8)
# tmp_img = tf.image.convert_image_dtype(tmp_img, dtype=tf.float32, saturate=True)
# tmp_img = tmp_img.numpy()
```

Apply the prediction mask, and the ground truth mask on the MRI image
and display the final output

```python
fig, axes = plt.subplots(1, 2, figsize=(6,6))
axes[0].imshow(tmp_img[..., 60], cmap="bone")
axes[0].imshow(tmp_seg[..., 60], cmap="Greens", alpha=0.5)
axes[0].axis("off")
axes[0].set_title("Predicted")

axes[1].imshow(tmp_img[..., 60], cmap="bone")
axes[1].imshow(seg_gt[..., 60], cmap="Greens", alpha=0.5)
axes[1].axis("off")
axes[1].set_title("True label")

plt.show()
```
